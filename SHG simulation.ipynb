```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

##constats:
wavelength=0.805 ##micrometers
n_0_w = 1.6603
n_e_w = 1.5454
n_0_2w=1.6924
n_e_2w=1.5676

theta_0=np.arcsin(np.sqrt((1/n_0_w**2-1/n_0_2w**2)/(1/n_e_2w**2-1/n_0_2w**2))) ## theta(0)
print(np.rad2deg(theta_0)) ## degrees print
l_0=200 ### micrometer units, our control paramater
##
alpha = np.arange(start=-np.pi/2,stop=np.pi/2,step=np.deg2rad(0.001)) ## x axis
print(len(alpha))
theta=theta_0+np.arcsin(np.sin(alpha)/n_0_w) ## theta(alpha)
l_alpha=l_0/(np.sqrt(1-(np.sin(alpha))**2/(n_0_w**2))) ## L(alpha)
dk_alpha=(2/wavelength)*(n_0_w - np.sqrt((n_e_2w**2 * n_0_2w**2)/(n_0_2w**2 * (np.sin(theta))**2 + n_e_2w**2 * (np.cos(theta)**2)))) ## the pi is included inside of np.sinc
intensity= (np.sinc(dk_alpha*l_alpha))**2 ## y axis
alpha_deg=np.rad2deg(alpha)
plt.plot(alpha_deg,intensity)
plt.xlabel(r'$ \alpha [rad] $')
plt.ylabel(r'$Normalize intensity [arb]$')
# plt.plot(alpha,np.linspace(start=0.5,stop=0.5,num=180000))
plt.show()

## We will expect to measure 40 < L_0 < 400 micrometer, otherwise we will get non physical solution.
```
